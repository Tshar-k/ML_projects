Linear regression of Perch fish weight


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/Perch_fish.csv")
```

```python
df
```

```python
x = df.kg
y = df.length
```

```python
plt.scatter(x,y,color = "blue",marker="+")
plt.xlabel("Weight",color = "green")
plt.ylabel("Length",color = "green")
```

```python
fish = linear_model.LinearRegression()
```

```python
fish.fit(df[['kg']],df.length)
```

```python
fish.predict([[500]])
```

```python
fish.coef_
```

```python
fish.intercept_
```

```python
plt.xlabel("Weight")
plt.ylabel("Length")
plt.scatter(x,y,color = "red",marker="+")
plt.plot(x,fish.predict(df[['kg']]))
plt.show()
```
