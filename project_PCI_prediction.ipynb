CANADA PER CAPITA INCOME PREDICTION (2017-2050)

```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
import pandas as pd
```

```python
df = pd.read_csv("/content/canada_per_capita_income.csv")
df.head(5)
```

```python
plt.scatter(df.year,df.income,color = "red",marker = '+')
plt.xlabel("year")
plt.ylabel("income (US$)")
plt.title("per capita income")
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(df[['year']],df.income)
```

```python
model.predict([[2020]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
d = pd.read_csv("/content/income_prediction.csv")
```

```python
d.head(5)
```

```python
p = model.predict(d)
```

```python
d["income"]= p
```

```python
print(d)
```

```python
d.to_csv("per_income_prediction.csv",index=False)
```
