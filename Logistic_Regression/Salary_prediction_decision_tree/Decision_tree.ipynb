```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
df = pd.read_csv("/content/Salaries.csv")
```

```python
df.head()
```

```python
inputs = df.drop(["salary_100k"],axis = "columns")
```

```python
target = df["salary_100k"]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_company = LabelEncoder()
label_degree = LabelEncoder()
label_job = LabelEncoder()
```

```python
inputs["company_n"] = label_company.fit_transform(inputs["company"])
```

```python
inputs["degree_n"] = label_company.fit_transform(inputs["degree"])
inputs["job_n"] = label_company.fit_transform(inputs["job"])
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis = 1)
```

```python
inputs_n
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
prediction = model.predict([[0,0,0]])
print(prediction)
```

```python
if (prediction == 1):
    print("Salary more than 100k")
else:
    print("Salary less than 100k")
```
