Outliers treatment

Outliers can create right skewness or leftskewness so to remove it log transformation can be done. The log transformation changes does not chnages the smaller value but it reduces the larger values

```python
train["LoanAmount_log"] = np.log(train["LoanAmount"])
sns.distplot(train["LoanAmount_log"])
```

```python
test["LoanAmount_log"] = np.log(test["LoanAmount"])
sns.distplot(test["LoanAmount_log"])
```
